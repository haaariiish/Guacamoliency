```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from rdkit.Chem.Scaffolds.MurckoScaffold import MurckoScaffoldSmiles
```

```python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
from rdkit.Chem.QED import qed
from rdkit.Chem.Crippen import MolLogP

IPythonConsole.ipython_useSVG=True 
```

```python
import sys
from io import StringIO
import os.path
```

```python
if not os.path.isfile("../data/processed/guacamol_scaffolds.csv") or not os.path.isfile("../data/processed/moses_scaffolds.csv"):
    original_moses_df = pd.read_csv("../data/raw/dataset_original_moses.csv")
    guacamol = pd.read_csv("../data/interim/guacamol.csv")
    original_moses_df["murcko_scaffolds_smiles"] = original_moses_df["SMILES"].apply(MurckoScaffoldSmiles)
    guacamol["murcko_scaffolds_smiles"] = guacamol["SMILES"].apply(MurckoScaffoldSmiles)
    original_moses_df.to_csv("../data/processed/moses_scaffolds.csv")
    guacamol.to_csv("../data/processed/guacamol_scaffolds.csv")
else :
    original_moses_df = pd.read_csv("../data/processed/moses_scaffolds.csv")
    guacamol = pd.read_csv("../data/processed/guacamol_scaffolds.csv")
```

```python
if not os.path.isfile("../data/processed/guacamol_complete.csv") or not os.path.isfile("../data/processed/moses_complete.csv"):
    original_moses_df["can_smiles"] = original_moses_df["SMILES"].apply(Chem.CanonSmiles)
    guacamol["can_smiles"] = guacamol["SMILES"].apply(Chem.CanonSmiles)
    original_moses_df.to_csv("../data/processed/moses_complete.csv")
    guacamol.to_csv("../data/processed/guacamol_complete.csv")
else:
    original_moses_df = pd.read_csv("../data/processed/moses_complete.csv")
    guacamol = pd.read_csv("../data/processed/guacamol_complete.csv")
```
