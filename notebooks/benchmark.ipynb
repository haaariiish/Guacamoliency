```python
true_max_position = info_saliency_moses['position of max score in saliency map']
difference = true_max_position - info_saliency_moses["position of first of a pair"]
difference = difference.to_list()
counting = dict()
for k in difference:
    if k in counting.keys():
        counting[k] +=1
    else :
        counting[k] =1

fig, ax = plt.subplots()
ax.bar(counting.keys(),counting.values(), alpha=0.6, label="moses canonical")
ax.set_ylabel('Occurences')
ax.set_xlabel('distance (by token)')
ax.set_title('distribution of distance between true max and first of a pair - token : ( )')
ax.legend()
ax.grid()
ax.plot()


true_max_position = info_saliency_moses['position of max score in saliency map']
difference = true_max_position - info_saliency_moses["position of second of a pair"]
difference = difference.to_list()
counting = dict()
for k in difference:
    if k in counting.keys():
        counting[k] +=1
    else :
        counting[k] =1

fig, ax = plt.subplots()
ax.bar(counting.keys(),counting.values(), alpha=0.6, label="moses canonical")
ax.set_ylabel('Occurences')
ax.set_xlabel('distance (by token)')
ax.set_title('distribution of distance between true max and second of a pair - token : ( )')
ax.legend()
ax.grid()
ax.plot()

difference = info_saliency_moses["position of second of a pair"] - info_saliency_moses["position of first of a pair"]
difference = difference.to_list()
counting = dict()
for k in difference:
    if k in counting.keys():
        counting[k] +=1
    else :
        counting[k] =1

fig, ax = plt.subplots()
ax.bar(counting.keys(),counting.values(), alpha=0.6, label="moses canonical")
ax.set_ylabel('Occurences')
ax.set_xlabel('distance (by token)')
ax.set_title('distribution of distance between first and second of a pair - token : ( )')
ax.legend()
ax.grid()   
ax.plot()
```

```python
n = 500
scores_precedent = info_saliency_moses["score of the token before the pair"]
scores_first = info_saliency_moses["score of first of a pair"]
```

```python
# Compute bins for both datasets
max1 = max(scores_precedent)
min1 = min(scores_precedent)
max2 = max(scores_first)
min2 = min(scores_first)

_min = min(min1, min2)
_max = max(max1,max2)

if _max >0.01:
    _max = 0.01

bin_width = (_max - _min) / n
x_bins = [_min + k * bin_width for k in range(n+1)]

# Create the plot
fig, ax = plt.subplots()

# Plot histograms
ax.hist(scores_precedent, bins=x_bins, alpha=0.6, label='score of the token before (',density=True)
ax.hist(scores_first, bins=x_bins, alpha=0.6, label='score of the first of a pair',density=True)
ax.grid()
ax.set_ylim((0,500))
ax.legend()
```

```python
scores_first_whenmax = info_saliency_guacamol[info_saliency_guacamol['position of max score in saliency map']- info_saliency_guacamol["position of first of a pair"] ==0]["score of first of a pair"]
scores_precedent_whenmax = info_saliency_guacamol[info_saliency_guacamol['position of max score in saliency map']- info_saliency_guacamol["position of first of a pair"] ==0]["score of the token before the pair"]

_max = max(max(scores_first_whenmax),max(scores_precedent_whenmax))
_min = min(min(scores_first_whenmax),min(scores_precedent_whenmax))
fig, ax = plt.subplots()
bin_width = (_max - _min) / n
x_bins = [_min + k * bin_width for k in range(n+1)]
ax.hist(scores_first_whenmax, bins=x_bins, alpha=0.6, label='score of the first of a pair',density=True)
ax.hist(scores_precedent_whenmax, bins=x_bins, alpha=0.6, label='score of the token before (',density=True)
ax.grid()
ax.set_ylim((0,10))
ax.legend()
```
