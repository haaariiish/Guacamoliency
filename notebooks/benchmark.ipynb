```python
import pandas as pd
from rdkit import Chem
import matplotlib.pyplot as plt
import numpy as np
```

```python
moses = pd.read_csv("../data/training_data/moses_canonical.csv")
```

```python
moses_smiles = moses["SMILES"].to_list()
```

```python
moses_mol = [Chem.MolFromSmiles(k) for k in moses_smiles]
```

```python
moses_inchies = [Chem.inchi.MolToInchi(k) for k in moses_mol if k]
```

# exploitation des donn√©es avec la saliency

```python
info_saliency_moses = pd.read_csv("../reports/data/moses_canonical_CL/1/pair_coherence_data_threshold0_sample_length13956.csv")
info_saliency_guacamol = pd.read_csv("../reports/data/guacamol_canonical_CL/1/pair_coherence_data_threshold0_sample_length7867.csv")
```

```python
def cut(s):
    s = s[1:len(s)-1]
    return int(s)
```

```python
true_max_position = info_saliency_moses['position of max score in saliency map'].apply(cut)
difference = true_max_position - info_saliency_moses["position of first of a pair"]
difference = difference.to_list()
counting = dict()
for k in difference:
    if k in counting.keys():
        counting[k] +=1
    else :
        counting[k] =1

fig, ax = plt.subplots()
ax.bar(counting.keys(),counting.values(), alpha=0.6, label="moses canonical")
ax.set_ylabel('Number per bar')
ax.set_xlabel('distance (by token)')
ax.set_title('distribution of distance between true max and first of a pair - token : ( )')
ax.legend()
ax.plot()


true_max_position = info_saliency_moses['position of max score in saliency map'].apply(cut)
difference = true_max_position - info_saliency_moses["position of second of a pair"]
difference = difference.to_list()
counting = dict()
for k in difference:
    if k in counting.keys():
        counting[k] +=1
    else :
        counting[k] =1

fig, ax = plt.subplots()
ax.bar(counting.keys(),counting.values(), alpha=0.6, label="moses canonical")
ax.set_ylabel('Number per bar')
ax.set_xlabel('distance (by token)')
ax.set_title('distribution of distance between true max and second of a pair - token : ( )')
ax.legend()
ax.plot()

difference = info_saliency_moses["position of second of a pair"] - info_saliency_moses["position of first of a pair"]
difference = difference.to_list()
counting = dict()
for k in difference:
    if k in counting.keys():
        counting[k] +=1
    else :
        counting[k] =1

fig, ax = plt.subplots()
ax.bar(counting.keys(),counting.values(), alpha=0.6, label="moses canonical")
ax.set_ylabel('Number per bar')
ax.set_xlabel('distance (by token)')
ax.set_title('distribution of distance between first and second of a pair - token : ( )')
ax.legend()    
ax.plot()
```
